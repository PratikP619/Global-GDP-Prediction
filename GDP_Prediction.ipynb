```python
# Importing the required Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
# Importing the Population dataset.

population_per_country=pd.read_csv("/content/PopulationPerCountry.csv",skiprows=range(0,4))
population_per_country.head()
```

```python
#importing GDP by GDP by countries.

GDP_country=pd.read_csv("/content/GDP by Country.csv",skiprows=range(0,4))
GDP_country.head()
```

**Merging the above 2 datasets**

```python
GDP_data=pd.merge(GDP_country,population_per_country,on=["Country Name","Country Code"], how="inner")
GDP_data.head(2)
```

```python
GDP_data.head()
```

```python
GDP_data1=pd.DataFrame()
for col in GDP_data.columns:
  if col.endswith("Name"):
    Country=col[:]
    GDP_data1[Country]=GDP_data[Country]
  if col.endswith("_x"):
      Year=col[:4]
      GDP_data1[Year]=GDP_data[Year+"_x"]/GDP_data[Year+"_y"]
  if col.endswith("Code"):
    Code=col[:]
    GDP_data1[Code]=GDP_data[Code]


GDP_data1.head()
```

```python
Model=["Multiple Linear Regression",
          "Polynomial Regression",
          "Decision Tree Regression",
          "Random Forest Regression"]


train_rmse={}
test_rmse={}
kfold={}
rsquared_train={}
rsquared_test={}
```

```python
# Splitting the data into X and Y

x=GDP.drop(["value"],axis=1)
x=GDP[["year","women_informed_choices","rural_population_percent","legal_rights","credit_to_private",
      "birthsattendedby_skilledstaff","ATM_machines","agriculture_machines","adult_literate","account_ratio"]]

y=GDP["value"]
print(x.shape,y.shape)
```

**Splitting data into Train Test**

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=40)
```

# **Model 1- Multiple Linear Regression**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model=LinearRegression()
model.fit(xtrain,ytrain)
print(model.intercept_)
print(model.coef_)
```

```python
# Multiple Linear Regression- Train

ypred_trainlr=model.predict(xtrain)
ytrain_act=ytrain
rmse_lr_train=np.sqrt(mean_squared_error(ypred_trainlr,ytrain_act))
rsq_train_lr=model.score(xtrain,ytrain)

print("rmse_lr_train",rmse_lr_train)
print("rsquared_trainLR",model.score(xtrain,ytrain))

train_rmse["Multiple Linear Regression"]=rmse_lr_train
rsquared_train["Multiple Linear Regression"]=rsq_train_lr
```

```python
ypred_trainlr=model.predict(xtrain)
ytrain_act=ytrain

mean_squared_error(ytrain_act,ypred_trainlr)
rmse_lr=np.sqrt(mean_squared_error(ytrain_act,ypred_trainlr))
print("rmse_lr:",rmse_lr)

rsq_train=model.score(xtrain,ytrain)
print("rsq_train:",model.score(xtrain,ytrain))

```

```python
# Multiple Linear Regression-Test

ypred_testlr=model.predict(xtest)

rmse_lr_test=np.sqrt(mean_squared_error(ytest,ypred_testlr))
rsq_lr_test=model.score(xtest,ytest)

print("rmse_lr_test:",rmse_lr_test)
print("rsquared_testLR",model.score(xtest,ytest))

test_rmse["Multiple Linear Regression"]=rmse_lr_test
rsquared_test["Multiple Linear Regression"]=rsq_lr_test
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

rmse_sum=0
cross_val=KFold(n_splits=10,random_state=None,shuffle=True)

for i in range (1,11):
  model_KF_LR=model.fit(x,y)
  score=cross_val_score(model_KF_LR,x,y,scoring="neg_mean_squared_error",cv=cross_val,n_jobs=1)
  rmse=np.sqrt(abs(score))
  rmse_sum=rmse_sum + np.mean(rmse)
  print("Linear Regression-"+str(i) +" MSE:" + str(np.mean(np.abs(score))) + ", STD: " + str(np.std(score)) + ", RMSE: " +str(np.mean(rmse)))
rmse_mean_lr = rmse_sum/10
print(rmse_mean_lr)
kfold["Multiple Linear Regression"] = rmse_mean_lr

```

```python
# Plotting actual VS predictes plot on test data

plot=sns.distplot(ypred_testlr,hist=True,bins=5,color="red")
plot=sns.distplot(ytest,hist=True,bins=5,color="blue")
plot.legend(labels=["Predicted","actual"])
plt.show()
```

# **Model 2- Polynomial**

```python
from sklearn.preprocessing import PolynomialFeatures

model_poly=PolynomialFeatures()
x_train_poly=model_poly.fit_transform(xtrain)
x_test_poly=model_poly.fit_transform(xtest)
model1=LinearRegression()
model1.fit(x_train_poly,ytrain)
```

```python
# Train-polynomial

ypred_train_poly=model1.predict(x_train_poly)
rmse_train_poly=np.sqrt(mean_squared_error(ytrain,ypred_train_poly))
rsq_train_poly=model1.score(x_train_poly,ytrain)

print("rmse_train_poly:",rmse_train_poly)
print("rsq_train_poly:",rsq_train_poly)

train_rmse["Polynomial Regression"]=rmse_train_poly
rsquared_train["Polynomial Regression"]=rsq_train_poly
```

```python
# Test-polynomial

ypred_test_poly=model1.predict(x_test_poly)
rmse_test_poly=np.sqrt(mean_squared_error(ytest,ypred_test_poly))
rsq_test_poly=model1.score(x_test_poly,ytest)


print("rmse_test_poly:",rmse_test_poly)
print("rsq_test_poly:",rsq_test_poly)

test_rmse["Polynomial Regression"]=rmse_test_poly
rsquared_test["Polynomial Regression"]=rsq_test_poly
```

```python
rmse=0
for i in range(1,11):
  poly=PolynomialFeatures(4)
  x_act=poly.fit_transform(x)
  model_kf_poly=model1.fit(x_act,y)
  score=cross_val_score(model_kf_poly,x,y,scoring="neg_mean_squared_error",cv=cross_val,n_jobs=1)
  rmse=np.sqrt(abs(score))
  rmse_sum=rmse_sum+np.mean(rmse)

  print("Polynomial   :"+str(i) +"Polynomial MSE:"+str(np.mean(np.abs(score))) + "STD:" +str(np.std(score)) +"rsme:" +str(np.mean(rmse)))
rmse_mean_poly=rmse_sum/10
print(rmse_mean_poly)
kfold["Polynomial Regression"]=rmse_mean_poly
```

```python
# Plotting Actual VS Predicted values of polynomial  Train data

plot=sns.distplot(ypred_train_poly,hist=True,bins=4,color="red")
plot=sns.distplot(ytrain,hist=True,bins=4,color="blue")

plot.legend(labels=["Predicted","Actual"] )
plt.show()
```
